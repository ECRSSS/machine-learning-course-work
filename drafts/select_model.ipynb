```python
merged_df = pd.read_csv('processed_data/prepared_featured_balanced_sample_3.csv').drop('Unnamed: 0',axis=1)

```

```python
merged_df = merged_df.sample(frac=1,random_state=42).reset_index()
merged_df
```

```python
merged_df = merged_df.drop(['index'],axis=1)
```

merged_df.dropna(subset = ['lat_mm_std'], inplace=True)
merged_df

```python
x = merged_df.drop(['event_target'],axis=1)
y = merged_df['event_target']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
def fit_and_result(classification_model):
    mlp = classification_model
    mlp.fit(x_train, y_train)

    mlp_train_pred = mlp.predict(x_train)
    mlp_test_pred = mlp.predict(x_test)

    print('Точность на тренировочной выборке')
    print(accuracy_score(y_train, mlp_train_pred))
    print('Точность на тестовой выборке')
    print(accuracy_score(y_test, mlp_test_pred))
    
    
    score = roc_auc_score(y_test, mlp_test_pred)
    print(f"\n\nROC AUC TEST: {score:.4f}")
```
