# Загружаем объединенные данные и готовим отбалансированный семпл по event_target (event_value)

```python
import pandas as pd
```

```python
TARGET_EVENTS =  ['sub_car_claim_click', 'sub_car_claim_submit_click',
'sub_open_dialog_click', 'sub_custom_question_submit_click',
'sub_call_number_click', 'sub_callback_submit_click', 'sub_submit_success',
'sub_car_request_submit_click']
```

```python
half_merged_df = pd.read_csv('processed_data/full_merged.csv')
```

```python
half_merged_df.columns.tolist()
```

```python
half_merged_df['event_action']
```

```python
half_merged_df.columns.tolist()
```

```python
def get_part_by_action(action):
    part = half_merged_df[half_merged_df['event_action'] == action]
    print(f'{action} = {len(part)} rows')
    if action in TARGET_EVENTS:
        return part
    else:
        return part.head(20000)

def get_balanced_sample():
    sample = half_merged_df.head(1)
    list_actions = half_merged_df['event_action'].unique().tolist()
    for action in list_actions:
        sample = pd.concat([sample,get_part_by_action(action)], axis=0)
    return sample
    


    
balanced_df = get_balanced_sample()
balanced_df
```

```python

balanced_df['event_target'] = balanced_df.apply(lambda row: row['event_action'] in TARGET_EVENTS,axis=1)
```

```python
balanced_df
```

```python
balanced_df['event_target'].value_counts()
```

```python
balanced_df = balanced_df.drop(['Unnamed: 0','event_category', 'event_action'],axis=1)

balanced_df
```

```python
balanced_df.event_target.value_counts()
```

```python
target_false_view_df = balanced_df[balanced_df.event_target == False].head(100000)
target_false_view_df.shape
```

```python
target_true_view_df = balanced_df[balanced_df.event_target == True]
target_true_view_df.shape
```

```python
balanced_merged_df = pd.concat([target_false_view_df.copy(),target_true_view_df.copy()], axis=0)
```

```python
balanced_merged_df
```

```python
balanced_df.to_csv('data/full_balanced_merged.csv')
```

```python
balanced_sessions_df = balanced_merged_df.drop(['hit_type', 'hit_date', 'hit_time', 'hit_number', 'hit_referer', 'hit_page_path', 'hit_referer'],axis=1).drop_duplicates()

balanced_sessions_df
```
